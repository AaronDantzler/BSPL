Run a regression of Weighted Score on dummy variables for the news stations

```python
y = df_combined['Weighted Score']
X = df_combined[['CNN', 'MSNBC', 'FOX']]
X = sm.add_constant(X)
lm = sm.OLS(y, X).fit()
```
