Using Logistic and Binary (1 or 0) Variables

```python
df_combined.info()
```

```python
df_combined["Y/N Score"] = df_combined["Y/N"].replace({'yes': 1, 'no': 0})
```

```python
df_combined
```

```python
# df_combined.loc[[0], ["Y/N"]]
```

```python
y = df_combined['Y/N Score']
X = df_combined[['CNN', 'MSNBC']]
X = sm.add_constant(X)
log = sm.Logit(y, X).fit()
log.summary()
```

```python
df_drop.info()
```

```python
df_drop["Sentiment Logit Score"] = df_drop['Sentiment'].replace({'risk': 0, 'opportunity': 1, 'neutral': 0})
```

```python
df_drop
```

```python
y = df_drop['Sentiment Logit Score']
X = df_drop[['CNN', 'MSNBC']]
X = sm.add_constant(X)
log = sm.Logit(y, X).fit()
log.summary()
```

```python
df_drop["Commitment Logit Score"] = df_drop['Commitment'].replace({'no': 0, 'yes': 1})
```

```python
y = df_drop['Commitment Logit Score']
X = df_drop[['CNN', 'MSNBC']]
X = sm.add_constant(X)
log = sm.Logit(y, X).fit()
log.summary()
```

```python
df_drop["Spec Logit Score"] = df_drop["Spec"].replace({'non': 0, 'spec': 1})
```

```python
y = df_drop['Spec Logit Score']
X = df_drop[['CNN', 'MSNBC']]
X = sm.add_constant(X)
log = sm.Logit(y, X).fit()
log.summary()
```
